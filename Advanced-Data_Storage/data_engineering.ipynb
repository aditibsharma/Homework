```python
# Import the Pandas library
import pandas as pd
import csv
```

```python
# load csv file
csv_path1 = '../Advanced-Data_Storage/hawaii_measurements.csv'
csv_path2 = '../Advanced-Data_Storage/hawaii_stations.csv'
measurements_df = pd.read_csv(csv_path1)
stations_df = pd.read_csv(csv_path2)
```

```python
measurements_df.dropna(axis=0, how='any')
stations_df.dropna(axis=0, how='any')
```

```python
measurements_df.to_csv('Clean_hawaii_measurements.csv')
stations_df.to_csv('Clean_hawaii_stations.csv')
```
